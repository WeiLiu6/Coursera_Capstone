```python
import pandas as pd
```

# Read and Clean data

```python
d = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M", flavor='bs4')
type(d)
```

```python
df = d[0]
df
```

```python
items = []
postcode = []
borough = []
neighb = []
rest=[]
for index, row in df.iterrows():
    for column in df.columns:
        items.append(row[column])
```

```python
for item in items:
    postcode.append(item[0:3])
    rest.append(item[3:])
```

```python
for i in rest:
    l = i.split('(')
    borough.append(l[0])
    if len(l)>1:
        neighb.append(l[1].replace(")", '').replace(' /', ','))
    else: neighb.append("Not assigned")
borough
```

```python
neighb
```

```python
can = {"PostalCode":postcode, "Borough":borough, "Neighborhood":neighb}
can_df = pd.DataFrame(data=can)
can_df
```

## Drop "Not assigned" Borough

```python
can_df = can_df[can_df['Borough']!='Not assigned']
can_df
```

## Set "Not assigned" Neighborhood value same as Borough

```python
can_df.loc[can_df['Neighborhood'] == 'Not assigned', 'Neighborhood'] = can_df.loc[can_df['Neighborhood']=='Not assigned', 'Borough']
can_df
```

```python
# Reset index
can_df.reset_index(drop=True, inplace=True)
can_df
```

```python
# Read the CSV with Latitude and Longitude
geo_df = pd.read_csv('Geospatial_Coordinates.csv')
geo_df
```

```python
# Merge two data frames
can_df = can_df.merge(geo_df, left_on='PostalCode', right_on="Postal Code")
can_df.drop(columns=['Postal Code'], inplace=True)
can_df
```

```python
toronto_df = can_df[can_df['Borough'].str.contains("Toronto")]
toronto_df.reset_index(drop=True, inplace=True)
toronto_df
```
